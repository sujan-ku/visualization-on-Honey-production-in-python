### QUESTION 3. Display the percentage distribution of the data in each year using the pie chart.

```python
YearCount=df["year"].value_counts()
```

```python
plt.figure(figsize=(10, 8))
plt.pie(YearCount, labels=YearCount.index, autopct="%1.1f%%", startangle=90)
plt.axis("equal") 
plt.title("Percentage Distribution of Data by Year")
plt.show()
```

### QUESTION4. Plot and Understand the distribution of the variable "price per lb" using displot, and write your findings.

```python

sns.distplot(df["priceperlb"], kde = False, bins = 5)
plt.show()
```

###### 
From to the above plot we can observe that the priceperlb is normally distributed.

```python
sns.displot(df, x="priceperlb", kde=True)
plt.show()
```

######

### QUESTION5. Plot and understand the relationship between the variables 'numcol' and 'prodval' through scatterplot, and write your findings.


```python
sns.scatterplot(data=df, x="numcol", y="prodvalue")
plt.title("Relationship between Number of Colonies and Production Value")
plt.xlabel("Number of Colonies")
plt.ylabel("Production Value")
plt.show()
```

######
From the above scatter plot we can see that the data is moving towards right in upper direction hence  there is a positive leniar relationship between "number of colonies" and "production value", hence we can say that increase in the number of colonies will increase the production valve

### QUESTION 6. Plot and understand the relationship between categorical variable 'year' and a numerical variable 'prodvalue' through boxplot, and write your findings.

```python
sns.boxplot(data=df, x="year", y="prodvalue")
plt.title("Relationship between Year and Production Value")
plt.xlabel("Year")
plt.ylabel("Production Value")
plt.show()
```

###### 
The above box plot shows outliers for each column concerning different year columns and their corresponding production valve.

### 7. Visualize and understand the relationship between the multiple pairs of variables throughout different years using pairplot and add your inferences. (use columns 'numcol', 'yield percol', 'total prod', 'prodvalue','year')


```python
columns = ["numcol", "yieldpercol", "totalprod", "prodvalue", "year"]
sns.pairplot(df[columns], hue="year", height=3)
plt.show()
```

###### 
from the above pairplot we can see some positive relationship they are:

######
*1)Between "numcol" and "total prod" we can see the positive linear relationship.As the number of colony increases ,the total production will also increase.
*2)Between "yield percol" and "total prod"we can see the positive linear relationship.As the yield per colony increases, the total production also increase.
*3)Between "total prod" and "prodvalue" we can see the positive linear relationship.As the total production increases, the production value also  increase.
*4)Between  "numcol" and "prodvalue"we can see the positive linear relationship.As the number of colonies increases, the production value also increase.
